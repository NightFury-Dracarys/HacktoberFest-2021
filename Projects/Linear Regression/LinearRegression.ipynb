```python
import seaborn as sb
data_corr = df1.corr()
sb.heatmap(data_corr,annot=True)
```

```python
from sklearn import linear_model
regress = linear_model.LinearRegression()

train_x = np.asanyarray(df1[['Y']])
train_y = np.asanyarray(df1[['area']])
#print(train_x)
#print(train_y)

regress.fit (train_x,train_y)
# The coefficients
print ('Coefficients: ', regress.coef_)
print ('Intercept: ',regress.intercept_)
```

```python
plt.scatter(df1.Y, df1.area,  color='blue',label="Actual")
plt.plot(train_x, regress.coef_[0][0]*train_x + regress.intercept_[0], '-r',label="Predicted")
plt.title("Area vs Y-spatial coord plot")
plt.xlim(2,9)
plt.ylim(0,1100)
plt.legend()
plt.xlabel("Y-spatial coord")
plt.ylabel("Area")
plt.show()
```

```python
y_predicted = regress.predict(train_x)
for i in range(0,len(train_x)):
    print(train_y[i],y_predicted[i])
df1['Predicted'] = y_predicted
print(y_predicted)
print(df1.head())
```
